```python
import os
from pathlib import Path

# Пути к файлам
data_folder = Path('./data/')
output_folder = Path('./output/')
output_folder.mkdir(exist_ok=True)

# Список файлов
files = os.listdir('data/')

all_text = []

for filename in files:
    filepath = data_folder / filename

    import re
    with open(filepath, 'r', encoding='utf-8') as f:
        content = f.read()

    # 1. Удаляем маркеры диалогов
    content = re.sub(r'^\s*[—\-–*]\s*', '', content, flags=re.MULTILINE)
    
    # 2. Переносы строк в пробелы
    content = re.sub(r'\n+', ' ', content)
    
    # 3. Удаляем английские буквы и цифры
    content = re.sub(r'[A-Za-z]', '', content)
    content = re.sub(r'\d+', '', content)
    content = re.sub(r'[^а-яА-ЯёЁ\s\.\,]', '', content)
    # 4. Удаляем все сочетания знаков (,., ,,, .,, и т.д.) целиком
    content = re.sub(r'[.,]{2,}', ' ', content)
    
    # 5. Удаляем знаки, которые висят отдельно между пробелами
    content = re.sub(r'\s+[.,]\s+', ' ', content)
    
    # 6. Удаляем знак в начале строки/текста
    content = re.sub(r'^[.,]+\s*', '', content)
    
    # 7. Удаляем знак в конце строки/текста
    content = re.sub(r'\s*[.,]+$', '', content)

    content = re.sub(r'\s+[.,]\s+', ' ', content)

    content = re.sub(r'\s+', ' ', content)
    content = re.sub(r'\s+[. ,]\s+', ' ', content)
    content = re.sub(r'\s+[, ,]\s+', ' ', content)
    content = re.sub(r'\s+', ' ', content)
    content = content.strip()

    if content:
        all_text.append(content.lower())


full_text = ' '.join(all_text)
print(f"Всего символов: {len(full_text):,}")

# Сохраняем
output_path = output_folder / 'tolstoy_cleaned.txt'
with open(output_path, 'w', encoding='utf-8') as f:
    f.write(full_text)
```
